```python
import numpy as np
from CompressTree import CompressTree
from numpy.linalg import svd
from TreeDraw import draw_tree
```

```python
n = 1024

X = np.random.random((n,n))
P = np.random.random((n,n))
P = (P > 0.999).astype(int)
X = P * X
X
```

```python
root = CompressTree(X, 0, len(X), 0, len(X[0]))
U, Sigma, V = svd(X)
Sigma
root.create_tree(16, Sigma[len(Sigma)-1])
```

```python
draw_tree(root)
```

```python
Y = np.zeros((n,n))
root.decompress(Y)
```

```python
root.compare(Y)
```
